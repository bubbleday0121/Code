## 使用句號(目前用這個)

```python
import os
import spacy
from textblob import TextBlob
import pandas as pd
from docx import Document
import re

# 載入 spaCy 的英文模型
nlp = spacy.load("en_core_web_lg")

# 定義資料夾路徑
folder_path = r"C:\Users\bubbl\OneDrive - 國立成功大學 National Cheng Kung University\NER\NFT press\Middle"

# 讀取 docx 檔案內容
def read_docx(file_path):
    doc = Document(file_path)
    full_text = []
    for para in doc.paragraphs:
        full_text.append(para.text)
    return "\n".join(full_text)

# 情感分析
def perform_sentiment_analysis(text):
    blob = TextBlob(text)
    sentiment = blob.sentiment.polarity
    if sentiment > 0:
        return "Positive", sentiment
    elif sentiment < 0:
        return "Negative", sentiment
    else:
        return "Neutral", sentiment

# 主題檢測（記錄觸發句、關鍵詞和情感，動態處理 AR/VR、Avatar 和 Luxury）
def detect_subject(text):
    text_lower = text.lower()
    doc = nlp(text)  # 使用 spaCy 分割句子
    sentences = [sent.text.strip() for sent in doc.sents]  # 獲取所有句子
    
    subjects = []
    nft_categories = {
        "Gaming": [r"\bgame\b", r"\bgaming\b", r"\bvideo game\b", r"\bvr\b", r"\bar\b", r"\baugmented reality\b", r"\bvirtual reality\b", r"\besports\b", r"\be-sports\b", r"\bavatar\b"],
        "Art": [r"\bartwork\b", r"\bartifacts\b", r"\bdigital art\b", r"\bart\b", r"\bgraphics\b", r"\bpainting\b", r"\bmeme\b"],
        "Collectibles": [r"\bcollectible\b", r"\bcollectibles\b", r"\bluxury\b"],
        "Music": [r"\bmusic\b", r"\bsong\b", r"\balbum\b"],
        "Metaverse": [r"\bmetaverse\b", r"\bvirtual world\b", r"\bvr\b", r"\baugmented reality\b", r"\bvirtual reality\b", r"\bavatar\b"],
        "Sports": [r"\bsport\b", r"\bsports\b", r"\bathlete\b", r"\bnba\b", r"\bnfl\b"],
        "Retail": [r"\bretail\b", r"\bluxury\b"],
        "Utility": [r"\bcoupon\b", r"\bticket\b", r"\bdomain name\b"],
        "Weather": [r"\bweather\b"],
        "Luxury": [r"\bluxury\b"],
        "Airline": [r"\bairline\b"],
        "Rent": [r"\brental\b"],
        "Others": [r"\basset ownership\b", r"\bownership\b"]
    }
    
    # 用來儲存每個類別的觸發句、關鍵詞和情感
    subject_sentences = {category: [] for category in nft_categories.keys()}
    subject_keywords = {category: set() for category in nft_categories.keys()}
    subject_sentiments = {category: [] for category in nft_categories.keys()}  # 儲存情感分數
    
    # 逐句檢查
    for sentence in sentences:
        sentence_lower = sentence.lower()
        for category, keywords in nft_categories.items():
            for keyword in keywords:
                if re.search(keyword, sentence_lower):
                    # 特別處理 AR/VR、Avatar 和 Luxury
                    special_keywords = [r"\bvr\b", r"\bar\b", r"\baugmented reality\b", r"\bvirtual reality\b", r"\bavatar\b", r"\bluxury\b"]
                    if keyword in special_keywords:
                        # 上下文檢查
                        has_game = any(re.search(g, sentence_lower) for g in [r"\bgame\b", r"\bgaming\b", r"\bvideo game\b", r"\besports\b"])
                        has_meta = any(re.search(m, sentence_lower) for m in [r"\bmetaverse\b", r"\bvirtual world\b"])
                        has_art = any(re.search(a, sentence_lower) for a in [r"\bartwork\b", r"\bartifacts\b", r"\bdigital art\b", r"\bart\b"])
                        has_collect = any(re.search(c, sentence_lower) for c in [r"\bcollectible\b", r"\bcollectibles\b"])
                        has_retail = re.search(r"\bretail\b", sentence_lower)
                        
                        if keyword in [r"\bvr\b", r"\bar\b", r"\baugmented reality\b", r"\bvirtual reality\b", r"\bavatar\b"]:
                            if has_game and not has_meta and not has_art and not has_collect:
                                if "Gaming" not in subjects:
                                    subjects.append("Gaming")
                                subject_sentences["Gaming"].append(sentence)
                                subject_keywords["Gaming"].add(keyword[2:-2] if keyword.startswith(r"\b") and keyword.endswith(r"\b") else keyword)
                                sentiment_label, sentiment_score = perform_sentiment_analysis(sentence)
                                subject_sentiments["Gaming"].append((sentiment_label, sentiment_score))
                            elif has_meta and not has_game and not has_art and not has_collect:
                                if "Metaverse" not in subjects:
                                    subjects.append("Metaverse")
                                subject_sentences["Metaverse"].append(sentence)
                                subject_keywords["Metaverse"].add(keyword[2:-2] if keyword.startswith(r"\b") and keyword.endswith(r"\b") else keyword)
                                sentiment_label, sentiment_score = perform_sentiment_analysis(sentence)
                                subject_sentiments["Metaverse"].append((sentiment_label, sentiment_score))
                            elif has_game and has_meta:
                                for cat in ["Gaming", "Metaverse"]:
                                    if cat not in subjects:
                                        subjects.append(cat)
                                    subject_sentences[cat].append(sentence)
                                    subject_keywords[cat].add(keyword[2:-2] if keyword.startswith(r"\b") and keyword.endswith(r"\b") else keyword)
                                    sentiment_label, sentiment_score = perform_sentiment_analysis(sentence)
                                    subject_sentiments[cat].append((sentiment_label, sentiment_score))
                            elif has_art or has_collect:
                                target = "Art" if has_art else "Collectibles"
                                if target not in subjects:
                                    subjects.append(target)
                                subject_sentences[target].append(sentence)
                                subject_keywords[target].add(keyword[2:-2] if keyword.startswith(r"\b") and keyword.endswith(r"\b") else keyword)
                                sentiment_label, sentiment_score = perform_sentiment_analysis(sentence)
                                subject_sentiments[target].append((sentiment_label, sentiment_score))
                            else:
                                if "Metaverse" not in subjects:
                                    subjects.append("Metaverse")
                                subject_sentences["Metaverse"].append(sentence)
                                subject_keywords["Metaverse"].add(keyword[2:-2] if keyword.startswith(r"\b") and keyword.endswith(r"\b") else keyword)
                                sentiment_label, sentiment_score = perform_sentiment_analysis(sentence)
                                subject_sentiments["Metaverse"].append((sentiment_label, sentiment_score))
                        elif keyword == r"\bluxury\b":
                            if has_retail and not has_art and not has_collect:
                                if "Retail" not in subjects:
                                    subjects.append("Retail")
                                subject_sentences["Retail"].append(sentence)
                                subject_keywords["Retail"].add("luxury")
                                sentiment_label, sentiment_score = perform_sentiment_analysis(sentence)
                                subject_sentiments["Retail"].append((sentiment_label, sentiment_score))
                            elif has_art and not has_retail and not has_collect:
                                if "Art" not in subjects:
                                    subjects.append("Art")
                                subject_sentences["Art"].append(sentence)
                                subject_keywords["Art"].add("luxury")
                                sentiment_label, sentiment_score = perform_sentiment_analysis(sentence)
                                subject_sentiments["Art"].append((sentiment_label, sentiment_score))
                            elif has_collect and not has_retail and not has_art:
                                if "Collectibles" not in subjects:
                                    subjects.append("Collectibles")
                                subject_sentences["Collectibles"].append(sentence)
                                subject_keywords["Collectibles"].add("luxury")
                                sentiment_label, sentiment_score = perform_sentiment_analysis(sentence)
                                subject_sentiments["Collectibles"].append((sentiment_label, sentiment_score))
                            elif has_retail or has_art or has_collect:
                                target = "Retail" if has_retail else "Art" if has_art else "Collectibles"
                                if target not in subjects:
                                    subjects.append(target)
                                subject_sentences[target].append(sentence)
                                subject_keywords[target].add("luxury")
                                sentiment_label, sentiment_score = perform_sentiment_analysis(sentence)
                                subject_sentiments[target].append((sentiment_label, sentiment_score))
                            else:
                                if "Luxury" not in subjects:
                                    subjects.append("Luxury")
                                subject_sentences["Luxury"].append(sentence)
                                subject_keywords["Luxury"].add("luxury")
                                sentiment_label, sentiment_score = perform_sentiment_analysis(sentence)
                                subject_sentiments["Luxury"].append((sentiment_label, sentiment_score))
                    else:
                        # 其他關鍵詞正常處理
                        if category not in subjects:
                            subjects.append(category)
                        subject_sentences[category].append(sentence)
                        subject_keywords[category].add(keyword[2:-2] if keyword.startswith(r"\b") and keyword.endswith(r"\b") else keyword)
                        sentiment_label, sentiment_score = perform_sentiment_analysis(sentence)
                        subject_sentiments[category].append((sentiment_label, sentiment_score))
    
    # 計算每個類別的平均情感（如果有多句）
    subject_avg_sentiments = {}
    for category in subjects:
        if subject_sentiments[category]:
            avg_score = sum(score for _, score in subject_sentiments[category]) / len(subject_sentiments[category])
            avg_label = "Positive" if avg_score > 0 else "Negative" if avg_score < 0 else "Neutral"
            subject_avg_sentiments[category] = (avg_label, avg_score)
        else:
            subject_avg_sentiments[category] = ("Neutral", 0.0)
    
    return subjects if subjects else ["General NFT"], subject_sentences, subject_keywords, subject_avg_sentiments

# 處理資料夾中的所有 docx 檔案
results = []
for filename in os.listdir(folder_path):
    if filename.endswith(".docx"):
        file_path = os.path.join(folder_path, filename)
        try:
            # 讀取檔案內容
            content = read_docx(file_path)
            
            # 檢測主題、觸發句、關鍵詞和情感
            subjects, subject_sentences, subject_keywords, subject_avg_sentiments = detect_subject(content)
            
            # 如果是 "General NFT"，記錄一行
            if subjects == ["General NFT"]:
                results.append({
                    "File": filename,
                    "Trigger_Sentence": "N/A",
                    "Subject": "General NFT",
                    "Keyword": "N/A",
                    "Sentiment": "Neutral",
                    "Sentiment_Score": 0.0,
                })
            else:
                # 為每個主題創建一行
                for subject in subjects:
                    trigger_sentences = "; ".join(subject_sentences[subject])
                    keywords = "; ".join(subject_keywords[subject])
                    sentiment_label, sentiment_score = subject_avg_sentiments[subject]
                    results.append({
                        "File": filename,
                        "Trigger_Sentence": trigger_sentences,
                        "Subject": subject,                        
                        "Keyword": keywords,
                        "Sentiment": sentiment_label,
                        "Sentiment_Score": sentiment_score,
                    })
            
            print(f"Processed: {filename}")
        except Exception as e:
            print(f"Error processing {filename}: {e}")

# 將結果轉為 DataFrame
df = pd.DataFrame(results)

# 計算每個 Subject 的情感平均值
sentiment_summary = df.groupby("Subject").agg(
    Average_Sentiment_Score=("Sentiment_Score", "mean"),
    Count=("Sentiment_Score", "count")
).reset_index()

# 根據平均情感分數添加情感標籤
sentiment_summary["Average_Sentiment"] = sentiment_summary["Average_Sentiment_Score"].apply(
    lambda x: "Positive" if x > 0 else "Negative" if x < 0 else "Neutral"
)

# 使用 ExcelWriter 將結果存成多個工作表
output_file = "new_nft_class_with_sentence_sentiment-middle.xlsx"
with pd.ExcelWriter(output_file, engine="openpyxl") as writer:
    # 第一個工作表：原始數據
    df.to_excel(writer, sheet_name="Detailed_Results", index=False)
    # 第二個工作表：每個 Subject 的情感平均值
    sentiment_summary.to_excel(writer, sheet_name="Sentiment_Summary", index=False)

print(f"Results saved to {output_file} with two sheets: 'Detailed_Results' and 'Sentiment_Summary'")
```
